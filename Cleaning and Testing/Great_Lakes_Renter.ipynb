```python
import pandas as pd
import numpy
from sklearn import preprocessing
```

```python
df = pd.read_csv('Resources\great_lakes_1.csv')
df.head()
```

```python
df = df.drop(['lat','long'],axis=1)
df.head()
```

```python
df.columns
category_columns = ['region','laundry_options','parking_options','type']
labelencoder = preprocessing.LabelEncoder()

mapping_dict = {}
for col in category_columns:
    df[col] = labelencoder.fit_transform(df[col])

    le_mapping = dict(zip(labelencoder.classes_,labelencoder.transform(labelencoder.classes_)))

    mapping_dict[col] = le_mapping

print(mapping_dict)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
           X_scaled, y, test_size = 0.3, random_state = 100)
```

```python
RFR = RandomForestRegressor(n_estimators=90)
RFR.fit(X_train,y_train)
y_pred = RFR.predict(X_test)

```

```python
from sklearn import metrics
import numpy as np
print('R^2:',metrics.r2_score(y_test, y_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_test, y_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
import pickle
filename = 'model.pkl'
pickle.dump(RFR,open(filename,'wb'))
```
